```python
df_csv = pd.read_csv('salescsv.csv')
df_excel=pd.read_excel('salesexcel.xlsx')
df_json = pd.read_json('salesjson.json')
```

```python
def clean_and_T(x):
    df_cleaned = x.dropna()
    if 'Date' in df_cleaned.columns:
        df_cleaned['Date']=pd.to_datetime(df_cleaned['Date'])

    df_cleaned.rename(columns={'Sales': 'Sales','region':'Region'},inplace = True)
    
    return df_cleaned
```

```python
df_csv_clean = clean_and_T(df_csv)
dfjson_clean = clean_and_T(df_json)
df_clean = clean_and_T(df_excel)
```

```python
df_combined = pd.concat([df_csv_clean,dfjson_clean,df_clean],ignore_index=True)
```
